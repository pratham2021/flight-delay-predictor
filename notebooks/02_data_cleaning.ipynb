# Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_parquet("../data/raw_data.parquet")
```

```python
df = df[df['DEP_DELAY'] <= 500]
```

```python
delay_cause_columns = ['LATE_AIRCRAFT_DELAY', 'SECURITY_DELAY', 'NAS_DELAY', 'WEATHER_DELAY', 'CARRIER_DELAY']
df[delay_cause_columns] = df[delay_cause_columns].fillna(0)
```

```python
df = df[df['CANCELLED'] == 0]
df = df[df['DIVERTED'] == 0]
```

```python
df = df.dropna(subset=['DEP_DELAY', 'ARR_DELAY', 'CRS_ELAPSED_TIME'])
```

```python
df.to_parquet('../data/cleaned_data.parquet', index=False)
```
