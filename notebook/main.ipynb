### 5.  Create IsHolidayPeriod using your holiday function

```python
def isHolidayPeriod(date):
    # df["MONTH"] = df["FL_DATE"].dt.month
    # df["DAY_OF_MONTH"] = df["FL_DATE"].dt.day
    
    # MLK Day
    if date.month == 1 and 15 <= date.day <= 20:
        return 1

    # Presidents Day
    if date.month == 2 and 13 <= date.day <= 17:
        return 1
    
    # Spring Break
    if (date.month == 3 and date.day >= 15) or (date.month == 4 and date.day <= 15):
        return 1
    
    # Memorial Day 
    if date.month == 5 and 23 <= date.day <= 27:
        return 1
    
    # Fourth of July
    if date.month == 7 and 1 <= date.day <= 7:
        return 1
    
    # Labor Day
    if (date.month == 8 and date.day >= 29) or (date.month == 9 and date.day <= 2):
        return 1
    
    # Thanksgiving (Nov 20 – Dec 1)
    if (date.month == 11 and date.day >= 20) or (date.month == 12 and date.day <= 1):
        return 1

    # Christmas & New Year (Dec 20 – Jan 5)
    if (date.month == 12 and date.day >= 20) or (date.month == 1 and date.day <= 5):
        return 1

    return 0
    
```

```python
df['IS_HOLIDAY_PERIOD'] = df['FL_DATE'].apply(isHolidayPeriod)
```

## Modeling

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix, roc_auc_score
```

```python
import lightgbm as lgbm
import optuna
```

```python
X_sample, _, y_sample, _ = train_test_split(X_train, y_train, train_size=0.1, stratify=y_train, random_state=1234)
X_s_train, X_s_val, y_s_train, y_s_val = train_test_split(X_sample, y_sample, test_size=0.2)
```

```python
training_data = lgbm.Dataset(X_s_train, label=y_s_train, params={'feature_pre_filter': False})
validation_data = lgbm.Dataset(X_s_val, label=y_s_val, reference=training_data, params={'feature_pre_filter': False})
```

### 1. Hyperparameter Tuning

```python
def objective(trial):
    params = {
        "objective": "binary",
        "metric": "auc",
        "verbosity": -1,
        "n_jobs": -1,
        "is_unbalance": True,
        
        "learning_rate": trial.suggest_float("learning_rate", 0.03, 0.10),
        "num_leaves": trial.suggest_int("num_leaves", 150, 300),
        "max_depth": trial.suggest_int("max_depth", 10, 14),
        "min_child_samples": trial.suggest_int("min_child_samples", 50, 200),
        "lambda_l1": trial.suggest_float("lambda_l1", 1e-3, 10.0, log=True),
        "lambda_l2": trial.suggest_float("lambda_l2", 1e-3, 10.0, log=True)
    }
    
    callbacks = [lgbm.early_stopping(stopping_rounds=30, verbose=False)]
    
    model = lgbm.train(
        params, 
        training_data, 
        valid_sets=[validation_data],
        num_boost_round=1000,
        callbacks=callbacks
    )
    
    return model.best_score["valid_0"]["auc"]
```

```python
study = optuna.create_study(direction="maximize")
```

```python
study.optimize(objective, n_trials=30)
```

### 2. Best Parameters
